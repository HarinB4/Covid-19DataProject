```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#read super Covid-19 dataframe with state as index. 
USCovid = pd.read_csv("../data/output/covid.csv")
Pop = pd.read_csv("https://covid.ourworldindata.org/data/ecdc/locations.csv")
AllCountries = pd.read_csv("https://covid.ourworldindata.org/data/ecdc/full_data.csv")
AllCountriesCases = pd.read_csv("https://covid.ourworldindata.org/data/ecdc/total_cases.csv")
AllCountriesDeaths = pd.read_csv("https://covid.ourworldindata.org/data/ecdc/total_deaths.csv")
USCovid = USCovid.drop(["Unnamed: 0","countyFIPS","stateFIPS"], axis=1)
```

```python
covid_df = USCovid.drop(USCovid.loc[USCovid['County Name'] == "Statewide Unallocated"].index)
sums = covid_df.select_dtypes(pd.np.number).sum(axis = 0).rename('UnitedStates')



```

```python
sums = sums.to_frame().transpose()
US_cases = sums.filter(regex = "x")
US_deaths = sums.filter(regex = "y")
```

```python
for i in range(1, len(US_cases.columns), 7):
    columns = [US_cases.columns[i - 1], US_cases.columns[i], US_cases.columns[i+1], 
           US_cases.columns[i+2],US_cases.columns[i+3], US_cases.columns[i+4], US_cases.columns[i+5]]   
    US_cases['Mean_week '+str(int((i/7)+1))] = US_cases[columns].mean(axis=1)
    US_cases['Median_week '+str(int((i/7)+1))] = US_cases[columns].median(axis=1)
```

```python
Five_CountriesCases = AllCountriesCases[["date","Bangladesh","Indonesia","Pakistan","Brazil","Nigeria","United States"]]
Five_CountriesDeaths = AllCountriesDeaths[["date","Bangladesh","Indonesia","Pakistan","Brazil","Nigeria","United States"]]
Five_CountriesCases[100:120]
```

```python
CasesMean = Five_CountriesCases.groupby(np.arange(len(Five_CountriesCases))//7).mean()
DeathsMean = Five_CountriesDeaths.groupby(np.arange(len(Five_CountriesDeaths))//7).mean()

```

```python
CasesMedian = Five_CountriesCases.groupby(np.arange(len(Five_CountriesCases))//7).median()
DeathsMedian = Five_CountriesDeaths.groupby(np.arange(len(Five_CountriesDeaths))//7).median()
```

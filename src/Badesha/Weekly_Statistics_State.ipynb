```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#read super Covid-19 dataframe with state as index. 
Read = pd.read_csv("../../data/output/covid.csv")
```

## Choose Texas as specific State 

```python
#groups the column by States, and adds up cases of each county per state. 
#array = ['TX']
#MyStates = Read.loc[Read['State'].isin(array)]
States = Read.groupby(['State']).sum()
States = States.drop(["countyFIPS","stateFIPS"], axis=1)
```

## Seperate Cases and Deaths of Specific State

```python
State_cases = States.filter(regex = "x").T.rename_axis('Date')
State_deaths = States.filter(regex = "y").T.rename_axis('Date')
```

```python
#Display State Cases
State_cases.head()
```

```python
#Display State Deaths
State_deaths.head()
```

## Calculate New Cases and New Deaths of States

```python
#Calculates differences of cases of each day (New Cases)
New_State_Cases = State_cases.diff()
New_State_Cases
```

```python
#Calculates differences of cases of each day (New Cases)
New_State_Deaths = State_deaths.diff()
New_State_Deaths
```

## Calculate weekly statistics

```python
#Calculate mean number of new cases per week
WeeklyNewCasesMean =  New_State_Cases.groupby(np.arange(len(New_State_Cases))//7).mean().astype(int).rename_axis('Week')
```

```python
WeeklyNewCases_Avg = WeeklyNewCasesMean.mean().to_frame().reset_index().round(1).rename(columns = {"0" : "Weekly Average"} )
WeeklyNewCases_Avg
```

```python
WeeklyNewCases_MED = WeeklyNewCasesMean.median().to_frame().reset_index().round(1)
WeeklyNewCases_MED
```

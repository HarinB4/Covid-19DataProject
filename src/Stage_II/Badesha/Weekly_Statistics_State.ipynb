```python
#read super Covid-19 dataframe with state as index. 
Read = pd.read_csv("../../../data/output/covid.csv")
df = Read.drop(["countyFIPS","stateFIPS","Unnamed: 0"], axis=1)
df = df.drop(df.loc[df['County Name'] == "Statewide Unallocated"].index)
```

## Get total amount of Cases and Deaths per state.

```python
#groups the column by States, and adds up cases of each county per state.
States = df.groupby(['State']).sum()
```

### Find States with similar population and compare them. 

```python
#Find states with similar Population.
#We choose Texas so we find other states with similar population(CA,Fl,NY, and PA)
States.sort_values(by=['population'],ascending = False)
```

## Choose States 

```python
array = ['TX','CA','FL','NY','PA','NC']
MyStates = States.loc[States.index.isin(array)]
MyStates
```

```python
#get Population value of States to use later on. 
TX_Pop = MyStates.iloc[5][0]
CA_Pop = MyStates.iloc[0][0]
FL_Pop = MyStates.iloc[1][0]
NC_Pop = MyStates.iloc[2][0]
NY_Pop = MyStates.iloc[3][0]
PA_Pop = MyStates.iloc[4][0]
```

## Seperate Cases and Deaths of Specific State

```python
State_cases = MyStates.filter(regex = "x").T.rename_axis('Date')
State_deaths = MyStates.filter(regex = "y").T.rename_axis('Date')
```

```python
#Display State Cases
State_cases.tail()
```

```python
#Display State Deaths
State_deaths.tail()
```

## Calculate New Cases and New Deaths of States

```python
#Calculates differences of cases and deaths of each day (New Cases & New Deaths)
New_State_Cases = State_cases.diff()
New_State_Deaths = State_deaths.diff()

print(New_State_Cases)
print(New_State_Deaths)

New_State_Cases.to_csv("Data/State_cases.csv")
New_State_Deaths.to_csv("Data/State_deaths.csv")
```

## Normalize by population (1 million)

```python
#calculate and Normalize dataset to cases per 1,000,000 people. 
norm_val = 1_000_000

#Normalize cases
Norm_MyStateCases = New_State_Cases.assign(CA = lambda x: (x['CA'] / CA_Pop) * norm_val).round(0)
Norm_MyStateCases = Norm_MyStateCases.assign(FL = lambda x: (x['FL'] / FL_Pop) * norm_val).round(0)
Norm_MyStateCases = Norm_MyStateCases.assign(NC = lambda x: (x['NC'] / NC_Pop) * norm_val).round(0)
Norm_MyStateCases = Norm_MyStateCases.assign(NY = lambda x: (x['NY'] / NY_Pop ) * norm_val).round(0)
Norm_MyStateCases = Norm_MyStateCases.assign(PA = lambda x: (x['PA'] / PA_Pop ) * norm_val).round(0)
Norm_MyStateCases = Norm_MyStateCases.assign(TX = lambda x: (x['TX'] / TX_Pop ) * norm_val).round(0)

#Normalize deaths 
Norm_MyStateDeaths = New_State_Deaths.assign(CA = lambda x: (x['CA'] / CA_Pop) * norm_val).round(0)
Norm_MyStateDeaths = Norm_MyStateDeaths.assign(FL = lambda x: (x['FL'] / FL_Pop) * norm_val).round(0)
Norm_MyStateDeaths = Norm_MyStateDeaths.assign(NC = lambda x: (x['NC'] / NC_Pop) * norm_val).round(0)
Norm_MyStateDeaths = Norm_MyStateDeaths.assign(NY = lambda x: (x['NY'] / NY_Pop ) * norm_val).round(0)
Norm_MyStateDeaths = Norm_MyStateDeaths.assign(PA = lambda x: (x['PA'] / PA_Pop ) * norm_val).round(0)
Norm_MyStateDeaths = Norm_MyStateDeaths.assign(TX = lambda x: (x['TX'] / TX_Pop ) * norm_val).round(0)

print(Norm_MyStateCases.tail())
print(Norm_MyStateDeaths.tail())                
```

### Changes date format for better split across the weeks. 

```python
#change date format for cases
MyStateCases  = Norm_MyStateCases.T.rename(columns=lambda x: (pd.to_datetime(x.split('_')[0], errors= 'ignore')), errors='raise').T
print(MyStateCases.tail())

#change date format for deaths
MyStateDeaths  = Norm_MyStateDeaths.T.rename(columns=lambda x: (pd.to_datetime(x.split('_')[0], errors= 'ignore')), errors='raise').T
print(MyStateDeaths.tail())

MyStateCases.to_csv("State_cases.csv")
MyStateDeaths.to_csv("State_deaths.csv")
```

## Calculate weekly statistics

```python
#Calculate mean number of new cases and New Deaths per week
WeeklyState_cases = MyStateCases.resample('W').mean().round(0).astype(int).rename_axis('Week')
WeeklyState_deaths = MyStateDeaths.resample('W').mean().round(0).astype(int).rename_axis('Week')


print(WeeklyState_cases)
print(WeeklyState_deaths)
```

```python
State_MeanCases = WeeklyState_cases.mean().round(0)
State_MedianCases = WeeklyState_cases.median()
ModeCases = WeeklyState_cases.mode()

State_MeanDeaths = WeeklyState_deaths.mean().round(0)
State_MedianDeaths = WeeklyState_deaths.median()
ModeDeaths = WeeklyState_deaths.mode()


State_ModeDeaths = ModeDeaths.values.tolist()
State_ModeCases = ModeCases.values.tolist()

```

```python
#Find population of the top 5 infected counties. 

array = ['Harris County','Dallas County','Bexar County','Tarrant County', 'Hidalgo County']

CountyPop = Texas.loc[Texas.index.isin(array)]

#get Population value of counties to use later on when normalizing. 
Harris_Pop = CountyPop.iloc[2][1]
Dallas_Pop = CountyPop.iloc[1][1]
Bexar_Pop = CountyPop.iloc[0][1]
Tarrant_Pop = CountyPop.iloc[4][1]
Hidalgo_Pop = CountyPop.iloc[3][1]
```

```python
Counties_Cases = Tex_cases.loc[Tex_cases.index.isin(array)].T.rename_axis('Date')
Counties_Deaths = Tex_deaths.loc[Tex_deaths.index.isin(array)].T.rename_axis('Date')
```

```python
#change date format for cases
Tex_Counties_Cases  = Counties_Cases.T.rename(columns=lambda x: (pd.to_datetime(x.split('_')[0], errors= 'ignore')), errors='raise').T
print(Tex_Counties_Cases.tail())

#change date format for deaths
Tex_Counties_Deaths  = Counties_Deaths.T.rename(columns=lambda x: (pd.to_datetime(x.split('_')[0], errors= 'ignore')), errors='raise').T
print(Tex_Counties_Deaths .tail())
```

```python
#check to see if dataset is correct
Tex_Counties_Cases.tail()
```

```python
#check to see if dataset is correct
Tex_Counties_Deaths.tail()
```

```python
#Calculates differences of cases and deaths of each day (New Cases & New Deaths)
New_Tex_Counties_Cases = Tex_Counties_Cases.diff()
New_Tex_Counties_Deaths = Tex_Counties_Deaths.diff()

print(New_Tex_Counties_Cases.tail())
print(New_Tex_Counties_Deaths.tail())
```

```python
#calculate and Normalize dataset to cases per 200,000 people. 
norm_val = 200_000

#Normalize cases
Norm_MyTxCases = New_Tex_Counties_Cases.assign(Harris_County = lambda x: (x['Harris County'] / Harris_Pop) * norm_val).round(0)
Norm_MyTxCases = Norm_MyTxCases.assign(Dallas_County = lambda x: (x['Dallas County'] / Dallas_Pop) * norm_val).round(0)
Norm_MyTxCases = Norm_MyTxCases.assign(Bexar_County = lambda x: (x['Bexar County'] / Bexar_Pop) * norm_val).round(0)
Norm_MyTxCases = Norm_MyTxCases.assign(Tarrant_County = lambda x: (x['Tarrant County'] / Tarrant_Pop) * norm_val).round(0)
Norm_MyTxCases = Norm_MyTxCases.assign(Hidalgo_County = lambda x: (x['Hidalgo County'] / Hidalgo_Pop) * norm_val).round(0)


#Normalize deaths 
Norm_MyTxDeaths = New_Tex_Counties_Deaths.assign(Harris_County = lambda x: (x['Harris County'] / Harris_Pop) * norm_val).round(0)
Norm_MyTxDeaths = Norm_MyTxDeaths.assign(Dallas_County = lambda x: (x['Dallas County'] / Dallas_Pop) * norm_val).round(0)
Norm_MyTxDeaths = Norm_MyTxDeaths.assign(Bexar_County = lambda x: (x['Bexar County'] / Bexar_Pop) * norm_val).round(0)
Norm_MyTxDeaths = Norm_MyTxDeaths.assign(Tarrant_County = lambda x: (x['Tarrant County'] / Tarrant_Pop) * norm_val).round(0)
Norm_MyTxDeaths = Norm_MyTxDeaths.assign(Hidalgo_County = lambda x: (x['Hidalgo County'] / Hidalgo_Pop) * norm_val).round(0)

MyTxCases = Norm_MyTxCases.drop(['Harris County','Dallas County','Bexar County','Tarrant County', 'Hidalgo County'], axis=1)
MyTxDeaths = Norm_MyTxDeaths.drop(['Harris County','Dallas County','Bexar County','Tarrant County', 'Hidalgo County'], axis=1)


print(MyTxCases.tail())
print(MyTxDeaths.tail())   
```

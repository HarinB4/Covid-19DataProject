# Trend calculation for the following states: MD,ME,MI,MN,MO,MS,MT,NC,ND,NE

```python
import pandas as pd
```

```python
last_day_date = '9/14/20'
label_cases = '_x'
label_death = '_y'
```

```python
#read the covid csv file 
covid_df = pd.read_csv('../data/output/new_covid.csv')
```

```python
#group all the county by state 
States = covid_df.groupby(['State']).sum()
```

## Get the 10 choosen states 

```python
#get a specific range of states 
State= States[20:30:]
State
```

## Get the trend for the cases in the last 7 days 

```python
#get the trend for cases in the 10 choosen states 
states_cases = State.columns.get_loc(last_day_date + label_cases)
cases_trend = State[State.columns[states_cases+(-6):states_cases+1]]
cases_trend

```

## For all the states, the number of confirmed cases is increasing every day 

## Get the trend for the deaths in the last 7 days

```python
#get the trend for death in the 10 choosen states 
states_death = State.columns.get_loc(last_day_date + label_death)
death_trend = State[State.columns[states_death+(-6):states_death+1]]
death_trend
```

### For all the states, the number of deaths is increasing every day except for ME state which remains constant.

## Covid_Employment_Merged

```python
#import libraries 
import pandas as pd
import numpy as np
```

## Read the enrichment dataset

```python
#reading the employment and covid dataset 
emp = pd.read_csv("../../data/Employment.csv", low_memory=False, thousands=',')
covid= pd.read_csv("../../data/output/covid.csv")
```

## Display the enrichment dataset 

```python
emp
```

### Dropping rows that do not match with the Covid dataset 

```python
#remove rows with US000 as FIPS , rows with MSA as area type and update the indices 
new_emp = emp.iloc[35:]
update= new_emp.iloc[:-6708]
update.index = np.arange(0, len(update))
update
```

### Rename columns and convert dataType to Int

```python
rename_emp = update.rename(columns={'Area\nCode':'countyFIPS'})
rename_emp["countyFIPS"] =rename_emp["countyFIPS"].astype(str).astype(int)
rename_emp["Establishment Count "] =rename_emp["Establishment Count"].astype(str).astype(int)
rename_emp["January Employment"] =rename_emp["January Employment"].astype(str).astype(int)
rename_emp["February Employment"] =rename_emp["February Employment"].astype(str).astype(int)
rename_emp["March Employment"] =rename_emp["March Employment"].astype(str).astype(int)
```

### Remove specific rows (unnecessary)

```python
indexname= rename_emp[rename_emp['Area Type'] == 'State'].index
rename_emp.drop(indexname, inplace=True)
```

### Select variables that will be used from employment dataset 

```python
column_drop = rename_emp[['countyFIPS', 'Establishment Count', 
                          'January Employment','February Employment','March Employment']]
column_drop
```

### Group all counties by the county FIPS and sum the number of employment 

```python
group_by_FIPS= column_drop.groupby(['countyFIPS']).sum()
group_by_FIPS
```

# Calculating Trends for Past Week in Georgia 

```python
georgia = covid.loc[covid['State'] == 'GA'].filter(regex='_x').iloc[:,-7:]
day = [x+1 for x in range(7)]
values = [ x for x in georgia.sum(axis=0)]
trend = pd.DataFrame()
trend['Day'] = day
trend['Case_Count'] = values
display(trend)
print(trend['Day'].corr(trend['Case_Count']))
```

# Trend Results: There is a strong positive correlation between Day and Case_Count, this means that as the week progressed, the number of COVID-19 cases increased.

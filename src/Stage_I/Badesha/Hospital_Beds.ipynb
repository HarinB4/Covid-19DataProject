# Display HospitalBeds Dataset and Merge with Super COVID-19 DataFrame

```python
import pandas as pd
import numpy as np

#Reads simplefied hospitalbed data and covid-19 dataset
read = pd.read_csv("../../../data/Hospital_Beds.csv")
supercovid = pd.read_csv("../../../data/output/covid.csv")
```

### Remove unnecessary variables, General display of Dataset. 

```python
#Displays Enrichment Data while removing unnecssary variables.
myhb = read.drop(["X","Y","HQ_ADDRESS","HQ_ADDRESS1","HQ_ZIP_CODE","HQ_CITY","CNTY_FIPS"], axis=1)

#renames columns so it will be easier to merge with supercovid-19 dataframe
myhb = myhb.rename(columns = {'HQ_STATE':'State','FIPS':'countyFIPS'})

#displays dataset to see progress
myhb
```

### Modify data into Hierarchical Indexes for easy readability and to merge with super COVID-19 Dataframe.

```python
#sets the Hierarchical index of states, counties, and countyFIPS
Hosbed = myhb.set_index(['STATE_NAME','COUNTY_NAME','countyFIPS'])

#removes more unnecessary variables that won't be used now. *Previous dataset is kept for hospital name purpose*
Hosbed = Hosbed.drop(["HOSPITAL_NAME","OBJECTID"], axis=1)

#sorts dataframe by FIPS and outputs
Hosbed = Hosbed.sort_values(by=['countyFIPS'])
Hosbed.head(10)
```

## Groups Hospital Bed dataset by counties of each state and sums values by each individual county. 

```python
#groups hospitalbed dataset by counties so integer variables can be added up together for easier merge with covid19 data
New_HosBed = Hosbed.groupby(['STATE_NAME','COUNTY_NAME','countyFIPS','STATE_FIPS']).sum()

#display dataset to see if groupby function works
New_HosBed.head(20)
```

```python
#merges the hospitalbed dataset with the covid19 data by countyFIPS which is the unique identifer
HosBedMerge = pd.merge(supercovid,New_HosBed, on = ["countyFIPS"], how = "left")

#drop unneccesarry extra columns that were added after the merge. 
HosBedMerge = HosBedMerge.drop(['Unnamed: 0'],axis=1)

#display if merge was a success. 
HosBedMerge
```

### Exports data to a csv file

```python
Merge_hb = HosBedMerge.to_csv("../../../data/output/COVID19_HOSBEDS_MERGE.csv")
```
